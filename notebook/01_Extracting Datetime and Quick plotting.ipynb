```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../data/dc_bikes.csv", parse_dates=["datetime"]) #loading DataFrame
```

### Extracting hour out of datetime

```python
df["hour"]= df["datetime"].dt.hour
df
```

### Extracting month out of datetime

```python
df["month"]= df["datetime"].dt.month
df
```

### Extracting dayofweek out of datetime (0 - Monday, 1 - Tuesday, ....)

```python
df["day"]= df["datetime"].dt.dayofweek
df
```

### Quick Plotting using pandas

```python
df.groupby("hour")["count"].count().plot(); #Number of rides per hour
```

```python
df.groupby("day")["count"].count().plot(); # number of rider per day of the week
```

```python
df.groupby("month")["count"].count().plot(); # number of rides per month
```

### Setting bin boundaries to form part of the day column

```python
df['hour'].min(), df['hour'].max()
```

```python
bin_labels = ['night','morning', 'afternoon', 'evening', 'night'] # setting bins
```

```python
bin_boundaries = [-1,4,12,16,20,23.1]
```

```python
df['part_of_day'] = pd.cut(df['hour'], bins=bin_boundaries, labels=bin_labels, ordered=False)

df
```

```python
df.to_csv("../data/dc_bikes_new.csv") #saving it locally
```

### Plotting with rolling average

```python
df_jan_2011 = df[(df["month"] == 1) & (df["datetime"].dt.year == 2011)] #filtering for Jan 2011
```

```python
df_jan_2011["time"] = df_jan_2011["datetime"].dt.strftime('%H:%M') #extracting time from datetime

df_jan_2011.head()
```

```python
df_jan_2011['rolling'] = df_jan_2011['count'].rolling(window=5).mean() #extracting rolling average with a window of 5
```

```python
df_jan_2011_plot = df_jan_2011.groupby("time").agg({"count":"sum","rolling":"sum"}) #aggregating based on time
```

```python
df_jan_2011_plot
```

```python
import matplotlib.pyplot as plt 
plt.rcParams['figure.dpi'] = 300

plt.plot(df_jan_2011_plot.index, df_jan_2011_plot['count'])
plt.plot(df_jan_2011_plot.index, df_jan_2011_plot['rolling'])
plt.xticks(rotation=45)
plt.title('Data With Rolling Average')

plt.show()
```
