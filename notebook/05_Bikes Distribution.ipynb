```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Loading Dataset

```python
df = pd.read_csv("../data/dc_bikes_new.csv",index_col=0, parse_dates=["datetime"])

df.head()
```

### Filtering for 07-01-2011

```python
df_day = df[df["datetime"].dt.strftime("%Y-%m-%d") == "2011-01-07"]

df_day.head()
```

### Distribution of Probability of number of rides

```python
sns.displot(df_day,x = 'count', kde= True,stat='probability', bins = 10)
```

### Filering for the year 2011

```python
df_year = df[df["datetime"].dt.year == 2011]
                      
df_year.head()
```

### Distribution of Probability of Temperature in 2011

```python
sns.displot(df_year,x = 'temp', kde= True,stat='probability', bins = 20)
```
